```python
import numpy as np
```

Shallow copy - view()  

```python
#  changes made to one array will reflect in the other
a = np.array([1,2,3,4])
b = a.view()
b[0] = 100
print(a)
```

```python
print(b)
```

Deep copy - copy()  - 

```python
# changes made to the copied array do not affect the original array
a = np.array([1,2,3,4])
c = a.copy()
c[0] = 100
print(a)
print(c)
```
