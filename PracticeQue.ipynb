```python
import numpy as np
```

```python
s = np.array([
    [5,3,4,6,7,8,9,1,2],
    [6,7,2,1,9,5,3,4,8],
    [1,9,8,3,4,2,5,6,7],

    [8,5,9,7,6,1,4,2,3],
    [4,2,6,8,5,3,7,9,1],
    [7,1,3,9,2,4,8,5,6],

    [9,6,1,5,3,7,2,8,4],
    [2,8,7,4,1,9,6,3,5],
    [3,4,5,2,8,6,1,7,9]
])
```

```python
b= np.sum(s,axis = 1)   # 1 for row sum
b
```

```python
b==45
```

```python
for i in b:
    print(i)
```

```python
for i in b:
    if i != 45:
        print("Sudoku is not valid")
        break
else:
    print("For rows it is valid")
```

```python
c = np.sum(s,axis = 0)   # 0 for column
for i in c:
    if i != 45:
        print("Sudoku is not valid")
        break
else:
    print("For column it is valid")
```

```python
import numpy as np
# columns: [Age, Math Marks, Science Marks]
data = np.array([
    [18,85,78],  # student 1
    [19,92,88],  # student 2
    [17,76,95],  #         3
    [18,65,70],  #         4
    [20,90, 85]  #         5
])
```

```python
# Get the shape of the matrix.
data.shape
```

```python
# Find the avg of the matrix.
np.mean(data[:,0])
```

```python
# Extract Math marks of all student
data[:,1]
```

```python
# Find the hiest science marks
np.max(data[:,2])
```

```python
#Get the details of the students who scored more than 90 in math.
data[data[:,1] >90]
```

```python
# increase Math marks of all students by 5
data[:,1] += 5
data
```

```python
# Find how many students are younger than  19
len(data[data[:,0]<19])
```

```python
#Calculate  the avg marks in each sub (column-wise maen)
np.mean(data[:,1:], axis = 0)
```

```python
# GET data of students who scored at least 80 in both subject.
data[(data[:,1] >= 80) & (data[:,2] >= 80)]
```

```python
# replace all science marks < 75 with 0
data[:,2][data[:,2] < 75] = 0
data
```
