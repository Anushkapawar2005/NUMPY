```python
import numpy as np
```

```python
# basic ex
a = np.array([10,20,30])
print(a.dtype)
```

```python
# Explicit datatype assignment
b = np.array([1,2,3], dtype='int32')
print(b.dtype)
```

```python
c = np.array([1.5,2.8,3.2])
print(c.dtype)
```

```python
d = np.array([2+3j, 4+5j])
print(d.dtype)
```

```python
e = np.array([True,False, True])
print(e.dtype)
```

```python
f = np.array(["Data","science","NumPy"])
print(f.dtype)   #  <U : unicode string     7: maximum length of characters
```

```python
g = np.array([10,3.5,"AI"], dtype=object) # heterogeneous data is unavoidable  because it is  shows last element datatype only
print(g.dtype)
```

```python
h = np.array([1,2,3])
print(h.itemsize)
i = h.astype(float)
print(i.dtype)
```

```python
arr = np.array([1,2,3,4,5], dtype = 'S')  # numpy converts num into bytes format
print(arr)        # b1 means sting '1' stored as byte
print(arr.dtype)   # |S1 - string byte , 1 - max length of each element , | - data stored in fixed memory
```

```python
arr = np.array([1,2,3,4], dtype= 'i4')
print(arr)
print(arr.dtype)
```

```python
arr = np.array(['a','2','3'], dtype = 'i')
print(arr)  # it gives error because a
```
