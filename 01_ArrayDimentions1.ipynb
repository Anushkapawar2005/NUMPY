```python
import numpy as np
```

```python
a = np.array([10,20,30,40,50,60])
print(a[3])
print(a[1:4])
```

```python
#using step
print(a[0:6:2])
```

```python
#negative slicing
print(a[-4:-1])
```

```python
b = np.array([[1,2,3],
               [4,5,6],
                [7,8,8]])
print(b)
```

```python
#slice specific rows
print(b[0:2, :])
```

```python
#slice specific columns
print(b[:,1:3])
```

```python
#slice sub-matrix
print(b[0:2,1:3])
```

```python
# step based slicing
print(b[::2,::2])
```

```python
c = np.array([
             [[1,2], [3,4]],
             [[5,6], [7,8]]
])
print(c)
```

```python
# slice first layer
print(c[0,:,:])
```

```python
print(c[1,0:2,0:1])
```

```python
# slice specific element
print(c[1,0,1])
```
