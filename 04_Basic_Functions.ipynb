```python
import numpy as np
```

Functions:

```python
a = np.array([[1,2,3], [4,5,6]])
print(a.shape)  # size of each dimention
```

```python
print(a.ndim)  # no of dimentions
```

```python
print(a.size)  # total count of elements across all dimentions
```

```python
print(a.dtype)  # which type of data stored
```

```python
b = np.array([1,2,3,4,5,6])
print(b.reshape(2,3))  # change the shape of array
```

```python
print(a.flatten())   # multidimentional array into single
```

Mathematical functions

```python
print(a.sum())
print(np.sum([1,2,3]))
```

```python
print(a.mean())
print(np.mean([10,20,30]))
```

```python
print(np.sort([4,1,3,2]))
c = np.array([6,2,1,9,3])
print(np.sort(c))
```

```python
# transpose()
print(a.T)
```

```python
a = np.array([10,40,20,60])
print(a.argmax())    # it returns index
print(a.argmin())
```

```python
# axis based operation
b = np.array([[1,2,3], [4,5,6]])
print(b.sum(axis = 0))  # column wise
print(b.sum(axis = 1))  # row wise
```

```python
c = np.array([1,2,2,3,4,4])
print(np.unique(c))
```

```python
d = np.array([10,30,50,70])
print(np.where(d>40, "High", "Low"))
```

```python
# clip() - defined minimum and max boundary
e = np.array([10,50,100])
print(np.clip(e,20,80))
```

```python
f = np.array([1.2,2.7,3.5])
print(np.round(f))  # nearest int
```

```python
print(np.floor(f))  # downwards to the nearest int
```

```python
print(np.ceil(f))   # upward to the nearest int
```
