```python
import numpy as np
```

Joining numpy array

```python
# joining two arrays using concatenate()
arr1 = np.array([1,2,3])
arr2 = np.array([4,5,6])
arr = np.concatenate((arr1,arr2))
print(arr)
```

```python
# joining two 2-D arrays along rows(axis = 1) using concatenate()  - row wise
arr1 = np.array([[1,2], [3,4]])
arr2 = np.array([[5,6], [7,8]])
arr = np.concatenate((arr1,arr2), axis = 1)
print(arr)
```

```python
arr1 = np.array([1,2,3])
arr2 = np.array([4,5,6])
arr = np.stack((arr1,arr2))
print(arr)      
```

```python
print(np.stack((arr1,arr2), axis = 1))  
```

```python
# stacking along rows - hstack()
print(np.hstack((arr1,arr2)))
```

```python
# stacking column wise - vstack()
print(np.vstack((arr1,arr2)))
```
